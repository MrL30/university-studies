# **Exercício 02**

## Clock arithmetic

Imagina que tens um relógio de 12 horas (AM/PM) ou de 24 horas, e queres calcular que horas serão após um certo número de horas. 

O teu objetivo é escrever um programa que:
* Pede ao utilizador para introduzir a hora atual (num formato de 12 ou 24 horas).
* Pede ao utilizador para introduzir um número de horas a adicionar.
* Usa o operador % para calcular e exibir a hora correta após adicionar essas horas, garantindo que o resultado se adapta ao formato do relógio escolhido.

Parte 1 – Relógio de 12 horas
Neste caso, o relógio deve "voltar ao início" após 12 horas. Por exemplo, se for 10 horas e adicionares 5 horas, o resultado será 3 horas (pois o relógio volta a contar a partir do 12).

Parte 2 – Relógio de 24 horas
Aqui, o relógio "volta ao início" após 24 horas. Por exemplo, se for 18 horas e adicionares 8 horas, o resultado será 2 horas (pois o relógio volta a contar após atingir 24).

Parte 3 – Escolha entre relógio de 12 ou 24 horas
Agora, podes combinar as duas partes anteriores num só programa, onde o utilizador escolhe se quer usar um relógio de 12 ou de 24 horas.

Instruções:
* O conceito de aritmética modular é útil neste exercício, onde o operador % ajuda a fazer o relógio "voltar ao início" após um certo número de horas.
* Experimenta com diferentes horas e valores para adicionar (ex.: começa às 11 horas e adiciona 5 ou 13 horas) para verificar se o programa está correto.

```python
# Exemplo de execução (Relógio de 12 horas):
Queres usar um relógio de 12 ou 24 horas? (escreve 12 ou 24): 12
Introduz a hora atual (formato 12 horas): 9
Quantas horas queres adicionar? 5
A hora será 2 horas após adicionar 5 horas.

# Exemplo de execução (Relógio de 24 horas):
Queres usar um relógio de 12 ou 24 horas? (escreve 12 ou 24): 24
Introduz a hora atual (formato 24 horas): 18
Quantas horas queres adicionar? 8
A hora será 2 horas após adicionar 8 horas.
```
