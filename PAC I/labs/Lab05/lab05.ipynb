- Para o pagamento com cartão ou cartão-presente, o troco não é necessário, mas o valor total da compra deve ser registado.

O programa deve permitir atender a vários clientes consecutivamente até que o operador escolha finalizar.

4. Exemplo de Funcionamento
- O programa solicita o valor total da compra do cliente.
- Pergunta pelo método de pagamento.
- Em caso de dinheiro: Solicita o valor pago pelo cliente e calcula o troco exato.
- Em caso de cartão ou cartão-presente: regista o valor da compra sem troco.
- Ao finalizar o atendimento de cada cliente, pergunta se deseja atender o próximo cliente.
- Ao encerrar o programa, exibe o relatório final.
