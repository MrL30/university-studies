# Lab09 - PAC 1

Este laboratório foca-se no uso de dicionários em Python. Complete os exercícios numerados de 1 a 10, implementando o código necessário nas células de código correspondentes.

## Exercício 1

Crie um dicionário chamado `inventario` com cinco produtos como chaves (strings) e as suas quantidades em stock como valores (inteiros). Imprima o dicionário completo e depois imprima apenas a quantidade de um dos produtos.

## Exercício 2

Dado o dicionário `temperaturas = {'Lisboa': 22, 'Porto': 18, 'Faro': 25, 'Braga': 19}`, adicione duas novas cidades com as suas temperaturas, remova uma cidade existente e atualize a temperatura de outra cidade. Imprima o dicionário final.

## Exercício 3

Crie um dicionário `contactos` onde as chaves são nomes de pessoas e os valores são dicionários com 'telefone' e 'email'. Adicione pelo menos três contactos e depois imprima apenas o email de um deles.

## Exercício 4

Escreva uma função `contar_palavras(texto)` que receba uma string e retorne um dicionário onde as chaves são as palavras únicas (em minúsculas) e os valores são o número de vezes que cada palavra aparece no texto. Teste com uma frase à sua escolha.

## Exercício 5

Crie um dicionário `produtos = {'P001': {'nome': 'Teclado', 'preco': 45.90}, 'P002': {'nome': 'Rato', 'preco': 25.50}, 'P003': {'nome': 'Monitor', 'preco': 199.99}}`. Escreva uma função `aplicar_desconto(produtos, codigo, percentagem)` que aplique um desconto percentual ao preço de um produto específico. Demonstre aplicando 10% de desconto a um produto.

## Exercício 6

Escreva uma função `fundir_dicionarios(dic1, dic2)` que receba dois dicionários e retorne um novo dicionário combinado. Se uma chave existir em ambos, o valor do segundo dicionário deve prevalecer. Teste com dois dicionários de configurações.

## Exercício 7

Crie um dicionário `vendas = {'Jan': {'produto1': 150, 'produto2': 200}, 'Fev': {'produto1': 180, 'produto2': 190}, 'Mar': {'produto1': 165, 'produto2': 210}}`. Escreva uma função `vendas_totais_por_produto(vendas)` que retorne um dicionário com o total de vendas de cada produto ao longo de todos os meses.

## Exercício 8

Defina uma classe `Estudante` que internamente use um dicionário `_atributos` para guardar nome, numero, curso e dicionário de notas por disciplina. Implemente métodos para adicionar notas a disciplinas, calcular a média geral, e obter todas as informações do estudante num formato de dicionário. Crie uma instância e demonstre todos os métodos.

## Exercício 9

Escreva uma função `agrupar_por_categoria(itens)` que receba um dicionário onde as chaves são nomes de itens e os valores são dicionários com 'categoria' e 'preco'. A função deve retornar um novo dicionário onde as chaves são categorias e os valores são dicionários com os itens dessa categoria. Teste com pelo menos 8 itens de 3 categorias diferentes.

## Exercício 10

Crie uma classe `SistemaReservas` que gerencie reservas de salas usando dicionários aninhados. A estrutura deve ser: {data: {sala: {hora: {'nome': str, 'contacto': str, 'participantes': int}}}}. Implemente os seguintes métodos: fazer_reserva(data, sala, hora, dados), cancelar_reserva(data, sala, hora), consultar_disponibilidade(data, sala) que retorna um dicionário com as horas disponíveis e ocupadas, listar_reservas_por_data(data) que retorna todas as reservas de um dia, e obter_estatisticas() que retorna um dicionário com o total de reservas por sala e a taxa de ocupação. Demonstre com pelo menos 6 reservas em diferentes datas, salas e horas, incluindo consultas, cancelamentos e estatísticas.
