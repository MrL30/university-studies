### Cálculo Estocástico de Pi usando o Método de Monte Carlo

**Contexto:**
O método de Monte Carlo permite estimar o valor de π através de simulações aleatórias. O princípio é o seguinte:

1. Imagine um quadrado com lado de comprimento 2, centrado na origem (de -1 a 1 em ambos os eixos)
2. Dentro deste quadrado, existe um círculo de raio 1, também centrado na origem
3. A área do quadrado é 4 (2×2)
4. A área do círculo é π×r² = π×1² = π
5. A razão entre a área do círculo e a área do quadrado é π/4

Se gerarmos pontos aleatórios dentro do quadrado e contarmos quantos caem dentro do círculo, a razão entre pontos dentro do círculo e total de pontos aproxima-se de π/4.

Um ponto (x, y) está dentro do círculo se: x² + y² ≤ 1

Portanto: **π ≈ 4 × (pontos dentro do círculo / total de pontos)**

A função principal deve receber o número de pontos a simular e retornar um tuplo com três elementos:
