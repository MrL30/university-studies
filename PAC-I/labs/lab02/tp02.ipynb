### Exercício 11 - Calculadora de Desconto

#### Objetivo
Neste exercício, vais praticar a recolha de inputs do utilizador e trabalhar com cálculos envolvendo percentagens. O programa deverá calcular o valor final de um produto após a aplicação de um desconto e também o valor do IVA (imposto sobre o valor acrescentado) que será aplicado.

#### Instruções
O programa deve pedir ao utilizador:
- O preço original de um produto.
- A percentagem de desconto a ser aplicada.
- A percentagem de IVA (Imposto sobre o Valor Acrescentado) a ser aplicada.

O programa deve calcular:
- O valor do desconto.
- O preço final após o desconto.
- O valor do IVA sobre o preço final.
- O preço final incluindo o IVA.

O programa deve então exibir os resultados.

#### Fórmulas:
Valor do Desconto: 
$ \text{Desconto}=\text{Preço Original} * \text{Percentagem de Desconto} / 100 $

Preço Final após Desconto: 
$\text{Preço Final}=\text{Preço Original} − \text{Desconto} $

Valor do IVA: 
$ \text{IVA}=\text{Preço Final} * \text{Percentagem de IVA} / 100 $

Preço Final com IVA: 
$ \text{Preço Final com IVA}=\text{Preço Final} + \text{IVA} $

Exemplo de Funcionamento Esperado:

```python
Insere o preço original do produto: 100
Insere a percentagem de desconto: 10
Insere a percentagem de IVA: 23

Valor do desconto: 10.0
Preço após desconto: 90.0
Valor do IVA: 20.7
Preço final com IVA: 110.7
```

#### Objetivo
Neste exercício, vais praticar a criação de variáveis e o uso de funções matemáticas para verificar uma regra da trigonometria. A regra diz que para qualquer valor de $x$, a soma dos quadrados do seno e do cosseno desse valor deve ser aproximadamente igual a 1. O objetivo é aplicar esta regra a um valor específico de $x$ e observar a diferença devido às limitações da aritmética de ponto flutuante.

#### Instruções
- Define uma variável chamada $x$ e atribui-lhe o valor de 60 (ou outro valor à escolha).
