# PAC01 - Laboratório Extra - Gestor de Passwords

## Objetivo
Implementar um sistema de gestão de passwords em Python que permita armazenar, recuperar e gerir passwords de forma segura para diferentes serviços.

## Requisitos

### Classes a Implementar

#### 1. `PasswordValidator`
Esta classe é responsável por validar a força das passwords.

**Métodos Obrigatórios:**
- `validate_password(password: str) -> bool`: Valida se uma password cumpre os requisitos de segurança
- `get_validation_errors(password: str) -> list`: Retorna uma lista de erros de validação

**Requisitos de Password:**
- Comprimento mínimo: 8 caracteres
- Deve conter pelo menos uma letra maiúscula (A-Z)
- Deve conter pelo menos uma letra minúscula (a-z)
- Deve conter pelo menos um dígito (0-9)
- Deve conter pelo menos um símbolo especial (!@#$%^&*()_+-=[]{}|;:,.<>?)

#### 2. `PasswordManager`
Esta classe gere o armazenamento e recuperação de passwords.

**Métodos Obrigatórios:**
- `__init__(master_password: str)`: Construtor que define a password mestra para o gestor
- `add_password(service: str, username: str, password: str) -> bool`: Adiciona uma nova entrada de password
- `get_password(service: str, master_password: str) -> str`: Recupera uma password (retorna password decifrada)
- `update_password(service: str, new_password: str, master_password: str) -> bool`: Atualiza uma password existente
- `delete_password(service: str, master_password: str) -> bool`: Remove uma entrada de password
- `list_services() -> list`: Retorna uma lista de todos os serviços armazenados
- `generate_password(length: int = 12) -> str`: Gera uma password aleatória segura

**Requisitos Adicionais:**
- As passwords devem ser cifradas antes de serem armazenadas
- A password mestra deve ser validada antes de qualquer operação
- Todas as passwords devem passar a validação antes de serem armazenadas
