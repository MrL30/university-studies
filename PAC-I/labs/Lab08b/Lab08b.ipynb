# Lab 08b - PAC 01



## Exercício 1
Cria uma classe base `Animal` com um método `speak()`. Cria duas classes derivadas, `Dog` e `Cat`, cada uma com a sua própria implementação do método `speak()`, que imprime sons típicos dos respetivos animais.

## Exercício 2
Cria uma classe `Funcionario` com um método que devolve o salário. Cria duas subclasses: `Professor` e `Tecnico`, que implementam o método para devolver salários diferentes.

## Exercício 3
Define uma classe estática `ConversorTemperatura` com dois métodos estáticos: um para converter Celsius para Fahrenheit, outro para converter Fahrenheit para Celsius.

## Exercício 4
Cria uma hierarquia de classes para veículos: uma classe base `Veiculo` e duas subclasses, `Carro` e `Moto`. Cada uma deve ter um método que apresenta o tipo de veículo.

## Exercício 5
Define uma classe `Figura` com um método `area()`. Cria as subclasses `Quadrado`, `Circulo`, `Retangulo` e `Triangulo` e implementa o método para cada uma.

## Exercício 6
Cria uma classe `BankAccount` com os seguintes requisitos:

1. Uma variável de classe para contar o número total de contas criadas.
2. Um método estático para validar se um saldo inicial é positivo.
3. Um método de classe para mostrar o número total de contas.
4. Um construtor que usa o método estático para validar o saldo inicial antes de criar a conta.
5. Um método para apresentar os valores da conta criada.

Crie na sua função `main` três contas bancárias diferentes. Verifique que os valores respeitantes às intâncias são efetivamente distintos entre as contas mas os valores da própria classe são partilhados entre as instâncias.
